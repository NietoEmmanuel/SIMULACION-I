Tendremos que la hipotesis nula es:
$$H_{0}= \mu = 0.5$$
Nuestra Hipotesis alterantiva será:
$$H_{0}= \mu \neq 0.5$$

Tendremos como estadisticos de prueba:

<center>$\overline{x}=\frac{x_1+x_2+...+x_N}{N}$

$Z_0=\frac{(\overline{x}-\frac{1}{2})\sqrt{N}}{\sqrt{\frac{1}{12}}}$ </center>

Y finalmente, si
$$|Z_o|<Z_{\frac{α}{2}}$$
Entonces no se rechazará la hipotesis
*   Debemos de recordar que $ α $ es nuestro nivel de confianza


```python
def prueba_promedio(datos, alpha):
  # n es el numero de datos que tiene el listado
  n = len(datos)

  # Promedio
  promedio = np.mean(datos)
  print ("El promedio de los datos es:", promedio)

  # Formulas para valor estadistico
  z0 = (promedio - 0.5) * np.sqrt(n) / np.sqrt(1/12)
  zC = stats.norm.ppf(1 - alpha/2)
  print ("El valor estadistico es:", abs(z0))
  print ("Vamos a comparar contra :", zC)
  # Validación de H0 o rechazo
  if abs(z0) < zC:
    print(f"Como {abs(z0):.4f} es menor que {zC:.4f} . No se rechaza la hipotesis nula H0")
  else:
    print("Se rechaza la hipotesis nula H0")

# Ingresar alpha
alpha = float(input("Ingrese el valor de alpha: "))

# Verificar que alpha este entre 0 y 1
if 0 < alpha < 1:
  # Se llama a lu función
  prueba_promedio(datos, alpha)
else:
  # Si  ponen otro valor de alpha, pide que metas un valor valido
  print("Ingrese algo entre 0 y 1")
```
