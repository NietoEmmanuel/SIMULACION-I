## **Uniforme**
* Nomenclatura: $X\sim U(0,1)$



*   Parámetros: $\mu, \sigma$
* Función de densidad de probabilidad: $f(x)=\frac{1}{b-a}$  si  $a\leq x \leq b$
* Función de distribución acumulada:$F(x)=\int_{-∞}^x f(u)du$
* **Media:** $E(x)=\frac{b+a}{2}$
* **Varianza:** $\sigma^2=\frac{(b-a)^2}{12}$
* **Desviación estándar:** $\sigma = \sqrt{\frac{(b-a)^2}{12}}$

```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace (a - 2, b + 2, 1000)
pdf = uniform.pdf(x, a, b - a)
cdf = uniform.cdf(x, a, b - a)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Uniforme Continua - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Uniforme Continua - Acumulativa")
plt.grid(True)
plt.show()

samples = np.random.uniform(a, b, 1000)
```

## **Triangular**
* Nomenclatura: $Tri(a,c,b)$



*   Parámetros: a *(límite inferior)*, b*(límite superior)*, c*(moda/pico)* donde $a\leq c \leq b$
* Función de densidad de probabilidad: $$ f(x) = \begin{cases} \frac{2(x-a)}{(b-a)(c-a)} & \text{para } a \leq x<c  \\ \frac{2}{b-a} & \text{para} \quad x = c \\ \frac{2(b-x)}{(b-a)(b-c)} & \text{para} \quad c<x\leq b \\ 0 & \text{para otros casos }  \end{cases} $$
* Función de distribución acumulada:$$ F(x) = \begin{cases} 0 & \text{para} \quad x\leq a \\ \frac{(x-a)^2}{(b-a)(c-a)} & \text{para } a < x\leq c  \\ 1- \frac{(b-x)^2}{(b-a)(b-c)} & \text{para} \quad c<x\leq b \\ 1 & \text{para } \quad b\leq x  \end{cases} $$
* **Media:** $E(x)=\frac{a+b+c}{3}$
* **Varianza:** $\sigma^2=\frac{a^2+ b^2 + c^2 - ab -ac-bc}{18}$
* **Desviación estándar:** $\sigma = \sqrt{\frac{a^2+ b^2 + c^2 - ab -ac-bc}{18}}$


```python
from scipy.stats import triang

a, b, c = 0, 10, 6
c_scaled = (c - a) / (b - a)
x = np.linspace(a, b, 1000)
pdf = triang.pdf(x, c_scaled, loc=a, scale=b - a)
cdf = triang.cdf(x, c_scaled, loc=a, scale=b - a)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Triangular - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Triangular - Acumulativa")
plt.grid(True)
plt.show()

samples = triang.rvs(c_scaled, loc=a, scale=b - a, size=1000)
```

## **Exponencial**
* Nomenclatura: $Exp(\lambda)$



*   Parámetros: $\lambda >0 $
* Función de densidad de probabilidad: $$ f(x) = \lambda e ^{-\lambda x}$$
* Función de distribución acumulada:$$ F(x) = 1- e^{-\lambda x} $$
* **Media:** $E(x)=\frac{1}{\lambda}$
* **Varianza:** $\sigma^2=\frac{1}{\lambda ^2 }$
* **Desviación estándar:** $\sigma = \frac{1}{\lambda}$

```python
from scipy.stats import expon

lam = 1
x = np.linspace(0, 10, 1000)
pdf = expon.pdf(x, scale=1/lam)
cdf = expon.cdf(x, scale=1/lam)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Exponencial - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Exponencial - Acumulativa")
plt.grid(True)
plt.show()

samples = np.random.exponential(1/lam, 1000)
```

## **m-Erlang**
* Nomenclatura: $Erlang(k, \lambda)$



*   Parámetros: $\lambda >0:$ *tasa de ocurrencia*

$k \in \mathbb{N}^+:$ *número de fases*
* Función de densidad de probabilidad: $$ f(x) = \frac{\lambda(\lambda x)^{n-1}e^{-\lambda x}}{(n-1)!}$$
* Función de distribución acumulada:$$ F(x) = 1 - \sum_{k=0}^{n-1} \frac{(\lambda x)^{n-1}e^{-\lambda x}}{k!} $$
* **Media:** $E(x)=\frac{k}{\lambda}$
* **Varianza:** $\sigma^2=\frac{k}{\lambda^2} $
* **Desviación estándar:** $\sigma = \sqrt{\frac{k}{\lambda^2}} $

```python
from scipy.stats import erlang

k, lam = 3, 1
x = np.linspace(0, 15, 1000)
pdf = erlang.pdf(x, k, scale=1/lam)
cdf = erlang.cdf(x, k, scale=1/lam)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución m-Erlang - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución m-Erlang - Acumulativa")
plt.grid(True)
plt.show()

samples = erlang.rvs(k, scale=1/lam, size=1000)
```

## **Gamma**
* Nomenclatura: $X\sim Gamma(α,β)$

* Función de densidad de probabilidad: $f(x) = (1 / (Γ(α) * β^α)) * x^{(α - 1)} * e^{(-x/β)}$

*   **Media:** αβ
*   **Varianza:** αβ²  
*   **Desviación estándar:** √(α) * β



```python
from scipy.stats import gamma

alpha, beta = 2, 2
x = np.linspace(0, 20, 1000)
pdf = gamma.pdf(x, alpha, scale=beta)
cdf = gamma.cdf(x, alpha, scale=beta)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Gamma - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Gamma - Acumulativa")
plt.grid(True)
plt.show()

samples = gamma.rvs(alpha, scale=beta, size=1000)
```

## **Weibull**
* Nomenclatura: $Weibull(k, \lambda )$

*   Parámetros: $k >0:$ *partámetro de forma*

$\lambda>0:$ *parámetro de escala*
* Función de densidad de probabilidad: $$ f(x) = \lambda \alpha (\lambda x)^{\alpha - 1}e^{(-\lambda x)\alpha}$$
* Función de distribución acumulada:$$ F(x) = 1- e^{-(\lambda x) \alpha}  $$
* **Media:** $E(x)=\lambda Τ (1+ \frac{1}{k})$
* **Varianza:** $V(x)= \lambda^2 [ Τ (1+\frac{2}{k}) - (Τ (1+ \frac{1}{k}))^2 ] $
* **Desviación estándar:** $D.E(x) = \sqrt{\lambda^2 [ Τ (1+\frac{2}{k}) - (Τ (1+ \frac{1}{k}))^2 ]} $

```python
from scipy.stats import weibull_min

k, lam = 1.5, 1
x = np.linspace(0, 5, 1000)
pdf = weibull_min.pdf(x, k, scale=lam)
cdf = weibull_min.cdf(x, k, scale=lam)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Weibull - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Weibull - Acumulativa")
plt.grid(True)
plt.show()

samples = weibull_min.rvs(k, scale=lam, size=1000)
```

## **Normal**
* Nomenclatura: $𝛮(\mu, \sigma^2 )$

* Función de densidad de probabilidad: $$ f(x) = \frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma ^2 }}$$
* Función de distribución acumulada:$$ F(x) = \frac{1}{\sigma \sqrt{2\pi}}\int_{-∞}^x e^{-\frac{(v-\mu)^2}{2\sigma ^2 }} dv $$
* **Media:** $E(x)=\mu$
* **Varianza:** $V(x)= \sigma^2 $
* **Desviación estándar:** $D.E(x) = \sigma $

```python

from scipy.stats import norm

mu, sigma = 0, 1
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='Densidad')
plt.title("Distribución Normal - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='Acumulativa', color="orange")
plt.title("Distribución Normal - Acumulativa")
plt.grid(True)
plt.show()
samples = np.random.normal(mu, sigma, 1000)
```

## **Log-normal**
* Nomenclatura: $LogNormal(\mu, \sigma)$



*   Parámetros: $\mu \in \mathbb{R}:$ *media normal subadyacente*

$\sigma>0:$ *desviación estándar de la normal subadyacente*
* Función de densidad de probabilidad: $$ f(x) = \frac{1}{\sigma x(\sqrt{2\pi})e^{-\frac{(lnx-\mu)^2}{2\sigma^2}}}$$
* Función de distribución acumulada:$$ F(x) = \Phi (\frac{lnx- \mu}{\sigma}) $$
* **Media:** $E(x)=e^{\mu+\frac{\sigma^2}{2}}$
* **Varianza:** $\sigma^2=(e^{\sigma^2}-1 )e^{2\mu+\sigma^2} $
* **Desviación estándar:** $\sigma = \sqrt{(e^{\sigma^2}-1 )e^{2\mu+\sigma^2}} $

```python
from scipy.stats import lognorm

mu, sigma = 0, 0.5
x = np.linspace(0.01, 5, 1000)
pdf = lognorm.pdf(x, sigma, scale=np.exp(mu))
cdf = lognorm.cdf(x, sigma, scale=np.exp(mu))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Log-normal - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Log-normal - Acumulativa")
plt.grid(True)
plt.show()

samples = lognorm.rvs(sigma, scale=np.exp(mu), size=1000)
```

## **Beta**
* Nomenclatura: $Beta(\alpha, \beta)$



*   Parámetros: $\alpha >0:$ *parámetro de forma*

$\beta:$ *parámetro de forma*
* Función de densidad de probabilidad: $$ f(x) = \frac{x^{\alpha -1(1-x)^{\beta -1}}}{B(\alpha, \beta)}$$
$B(\alpha, \beta)$ es la función Beta.
* Función de distribución acumulada:$$ F(x) = \frac{B(x;\alpha, \beta)}{B(\alpha, \beta)}=I_x(\alpha,\beta) $$
$I_x(\alpha,\beta)$ es la función Beta incompleta regularizada.
* **Media:** $E(x)=\frac{\alpha}{\alpha + \beta} $
* **Varianza:** $\sigma^2=\frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)} $
* **Desviación estándar:** $\sigma = \sqrt{\frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}} $

```python
from scipy.stats import beta

a, b = 2, 5
x = np.linspace(0, 1, 1000)
pdf = beta.pdf(x, a, b)
cdf = beta.cdf(x, a, b)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Beta - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Beta - Acumulativa")
plt.grid(True)
plt.show()
samples = beta.rvs(a, b, size=1000)
```

## **Pearson (Tipos)**

La familia de distribuciones Pearson sirve para modelar variables aleatorias con diferentes grados de asimetría y curtosis. Estas distribuciones están definidas como soluciones a la siguiente ecuación diferencial:

$$ \frac{1}{f(x)} \cdot \frac{df(x)}{dx} = - \frac{a_1 + a_2 x}{b_0 + b_1 x + b_2 x^2} $$

donde $f(x)$ es la función de densidad de probabilidad, y los coeficientes $ a_1, a_2, b_0, b_1, b_2 $ determinan la forma específica de la distribución.

Según la naturaleza del polinomio cuadrático en el denominador, se obtienen distintas formas de distribución, clasificadas en ocho tipos principales:

**Tipo I (Beta generalizada)**
$$f(x) = \frac{(x - a)^{m - 1} (b - x)^{n - 1}}{(b - a)^{m + n - 1} B(m, n)}, \quad a < x < b$$

**Tipo II (Beta simétrica especial)**
$$f(x) = \frac{(1 - x^2)^{m - 1}}, \quad -1 < x < 1$$

**Tipo III (Gamma)**
$$f(x) = \frac{x^{k - 1} e^{-x/\theta}}{\theta^k \Gamma(k)}, \quad x > 0$$

**Tipo IV**
$$f(x) = C \cdot \left(1 + \left(\frac{x - \mu}{\sigma}\right)^2\right)^{-m} \cdot \exp\left[\nu \cdot \tan^{-1}\left(\frac{x - \mu}{\sigma}\right)\right]$$

**Tipo V (Inversa Gamma)**
$$f(x) = \frac{\beta^\alpha}{\Gamma(\alpha)} x^{-\alpha - 1} e^{-\beta/x}, \quad x > 0$$

**Tipo VI (Beta invertida)**
$$f(x) = \frac{x^{p - 1} (1 + x)^{-(p + q)}}{B(p, q)}, \quad x > 0$$

**Tipo VII (t-Student generalizada)**
$$f(x) = C \cdot \left(1 + \left(\frac{x - \mu}{\sigma}\right)^2\right)^{-m}$$

**Normal**
$$f(x) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

*   Funciones de distribución

**Tipo I (Beta generalizada)**
$$F(x) = I_{\frac{x - a}{b - a}}(m, n), \quad \text{para } x \in (a, b)$$

**Tipo II (Beta simétrica especial)**

## **Johnson**
* Nomenclatura: $ z \sim \mathcal{N}(0, 1) $

* Función de densidad de probabilidad.
Todos los tipos comparten la forma:
$$z = \gamma + \delta \cdot g\left( \frac{x - x_i}{\lambda} \right)
$$

*   Parámetros:
donde:

$ z \sim \mathcal{N}(0, 1) $

$\gamma \in \mathbb{R} $: parámetro de forma

 $ \delta > 0 $: parámetro de forma

$ x_i \in \mathbb{R} $: parámetro de ubicación

 $ \lambda > 0 $: parámetro de escala

 $ g(\cdot) $: transformación específica del tipo de distribución Johnson



 $λ>0:$ tasa media de ocurrencias por unidad de tiempo/espacio



* Función de distribución acumulada:

La función de distribución acumulada para cualquier tipo de distribución de la familia Johnson se expresa como:

$$
F_X(x) = \Phi\left( \gamma + \delta \cdot g\left( \frac{x - \xi}{\lambda} \right) \right)
$$

donde:
 $ \Phi(\cdot) $: función de distribución acumulada de la normal estándar

 $\gamma \in \mathbb{R} $: parámetro de forma.

$ \delta > 0 $: parámetro de forma.

 $ \xi \in \mathbb{R} $: parámetro de ubicación.

$\lambda > 0 $: parámetro de escala.

$ g(\cdot) $: transformación específica dependiendo del tipo de distribución Johnson.



* **Media, varianza y desv.** calculados por medio de código, ya que de forma analitica es más complejo.

```python
from scipy.stats import johnsonsu

gamma_param, delta_param, xi, lambd = 1.5, 1.2, 0, 1
x = np.linspace(-5, 5, 1000)
pdf = johnsonsu.pdf(x, gamma_param, delta_param, loc=xi, scale=lambd)
cdf = johnsonsu.cdf(x, gamma_param, delta_param, loc=xi, scale=lambd)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Johnson SU - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Johnson SU - Acumulativa")
plt.grid(True)
plt.show()

samples = johnsonsu.rvs(gamma_param, delta_param, loc=xi, scale=lambd, size=1000)
```

## **Discreta Uniforme**
* Nomenclatura: $X\sim U(a,b), a\leq b$

*   Parámetros: $a:$ valor mínimo, $b:$ valor máximo
* Función de densidad de probabilidad: $P(X=x)=\frac{1}{n}$
* Función de distribución acumulada:$F(k;a,b)=\frac{[k]-a+1}{b-a+1}$
* **Media:** $E(x)=\frac{a+b}{2}$
* **Varianza:** $\sigma^2=\frac{(b-a+1)^2-1}{12}$
* **Desviación estándar:** $\sigma = \sqrt{\frac{(b-a+1)^2-1}{12}}$


```python
a, b = 1, 6
x = np.arange(a, b + 1)
pmf = [1/(b - a + 1)] * len(x)
cdf = np.cumsum(pmf)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
# Removed the deprecated 'use_line_collection' argument
plt.stem(x, pmf)
plt.title("Uniforme Discreta - PMF")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post', color='orange')
plt.title("Uniforme Discreta - CDF")
plt.grid(True)
plt.show()

samples = np.random.randint(a, b+1, 1000)
```

## **Poisson**
* Nomenclatura: $Poisson(\lambda)$



*   Parámetros: $λ>0:$ tasa media de ocurrencias por unidad de tiempo/espacio


* Función de densidad de probabilidad: $$P(X = x) = \frac{e^{-\lambda} \lambda^x}{x!}, \quad x = 0, 1, 2, \dots
$$
* Función de distribución acumulada:$$P(X \leq x) = \sum_{k=0}^{x} \frac{e^{-\lambda} \lambda^k}{k!}$$


* **Media:** $E(x)=\mu$
* **Varianza:** $\sigma^2=\mu$
* **Desviación estándar:** $\sigma = \sqrt{\mu}$

```python
from scipy.stats import poisson

lam = 3
x = np.arange(0, 15)
pmf = poisson.pmf(x, lam)
cdf = poisson.cdf(x, lam)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
# Removed the deprecated 'use_line_collection' argument
plt.stem(x, pmf)
plt.title("Distribución de Poisson - PMF")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post', color='orange')
plt.title("Distribución de Poisson - CDF")
plt.grid(True)
plt.show()

# Simulación
samples = np.random.poisson(lam, 1000)
```

## **Bernoulli**
* Nomenclatura: $Bernoulli(p)$



*   Parámetros: $p \in (0,1):$ probabilidad de éxito


* Función de densidad de probabilidad: $P(X=x)=p^x(1-p)^{1-x}$
* Función de distribución acumulada:$F(x) = \begin{cases} 0 & \text{para} \quad x<0  \\ 1-p & \text{para } 0 \leq x< 1  \\ 1& \text{para} \quad x\geq 1  \end{cases} $
* **Media:** $E(x)=p$
* **Varianza:** $\sigma^2=p(1-p)$
* **Desviación estándar:** $\sigma = \sqrt{p(1-p)}$

```python
from scipy.stats import bernoulli


p = 0.3
x = [0, 1]
pmf = bernoulli.pmf(x, p)
cdf = bernoulli.cdf(x, p)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title("Bernoulli")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post', color='orange')
plt.title("Bernoulli - Función Acumulativa")
plt.grid(True)
```

## **Geométrica**
* Nomenclatura: $Geom(p)$



*   Parámetros: $p \in (0,1]:$ probabilidad de éxito


* Función de densidad de probabilidad: $P(X=x)=p(1-p)^{1-x}$

* **Media:** $E(x)=\frac{1}{p}$
* **Varianza:** $\sigma^2=\frac{1-p}{p^2}$
* **Desviación estándar:** $\sigma = \sqrt{\frac{1-p}{p^2}}$

```python
from scipy.stats import geom

p = 0.4
x = np.arange(1, 15)
pmf = geom.pmf(x, p)
cdf = geom.cdf(x, p)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title("Geométrica - Función de Masa")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post', color='orange')
plt.title("Geométrica - Función Acumulativa")
plt.grid(True)
```

## **Binomial negativa**
* Nomenclatura: $NegBin(r,p)$



*   Parámetros: $r>0:$  número de éxitos deseados

$p \in (0,1]:$ probabilidad de éxito


* Función de densidad de probabilidad: $P(X = x) = \binom{x - 1}{r - 1} p^r (1 - p)^{x - r}, \quad x = r, r+1, r+2, \dots$
* Función de distribución acumulada:$$P(X \leq x) = \sum_{k=r}^{x} \binom{k - 1}{r - 1} p^r (1 - p)^{k - r}$$


* **Media:** $E(x)=\frac{r}{p}$
* **Varianza:** $\sigma^2=\frac{r(1-p)}{p^2}$
* **Desviación estándar:** $\sigma = \sqrt{\frac{r(1-p)}{p^2}}$

```python
from scipy.stats import nbinom
r, p = 5, 0.4
x = np.arange(0, 30)
pmf = nbinom.pmf(x, r, p)
cdf = nbinom.cdf(x, r, p)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title("Binomial Negativa - Función de Masa")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post', color='orange')
plt.title("Binomial Negativa - Función Acumulativa")
plt.grid(True)
```
