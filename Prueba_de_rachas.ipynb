<h1 style="text-align: center">PRUEBAS DE RACHAS </h1>

Para esta prueba se considera la siguiente hipótesis nula:
$$H_0 : \text{Los números de la muestra } \{X_i\} \text{ son independientes}$$
Luego,
$$H_1 : \text{Los números de la muestra } \{X_i\} \text{ no son independientes}$$
---
## Paso 1: Determine la secuencia
Determine la secuencia $S$ de $0$ y $1$ (o $+$ y $-$) de la siguiente manera:
- Si $X_i \leq X_{i-1}$ escribir **0** (o **+**)
- Si $X_i > X_{i-1}$ escribir **1** (o **−**)
La secuencia $S$ contiene $n-1$ números.

Ejemplo:
$$S = \{-, -, +, -, +, +, +, -, -, +, -, +\}$$
---
## Paso 2: Contar el número de rachas

Contar el número de rachas observadas $C_o$.

Una **racha** viene dada por el número de ceros o unos consecutivos que la forman.

---

## Paso 3: Calcular valores esperados
Calcule el valor esperado y la varianza del número de rachas, así como el estadístico $Z_0$.

Media:
$$\mu_{C_o} = \frac{2n - 1}{3}$$
Varianza:
$$\sigma_{C_o}^{2} = \frac{16n - 29}{90}$$
Estadístico de prueba:
$$Z_0 = \frac{C_o - \mu_{C_o}}{\sigma_{C_o}}$$
---
Dado que $C_o$ sigue aproximadamente una distribución normal con media $\mu_{C_o}$ y varianza $\sigma_{C_o}^2$, entonces
$$Z_0 \sim N(0,1)$$

---

## Región de aceptación

El intervalo de aceptación es:

$$- Z_{\alpha/2} \leq Z_0 \leq Z_{\alpha/2}$$

```python
def prueba_rachas(datos, alpha=0.05):

    n = len(datos)
    # Creamos la secuencia
    secuencia = []
    for i in range(1, n):
        if datos[i] > datos[i-1]:
            secuencia.append("+")
        else:
            secuencia.append("-")

    # Contaremos las Rachas
    c0 = 1
    for i in range(1, len(secuencia)):
        if secuencia[i] != secuencia[i-1]:
            c0 += 1


    # Calculamos el estadistico
    mu = (2*n - 1) / 3
    var = (16*n - 24) / 90
    sigma = math.sqrt(var)
    z0 = (c0 - mu) / sigma

    z_critico = norm.ppf(1 - alpha/2)

    if -z_critico <= z0 <= z_critico:
        decision = "No se rechaza H0 (Hay independencia)"
    else:
        decision = "Se rechaza H0 (No hay independencia)"

    print("Secuencia:", secuencia)
    print("Número de rachas:", c0)
    print("Media esperada:", mu)
    print("Varianza:", var)
    print("Z0 calculado:", z0)
    print("Z crítico:", z_critico)
    print("Decisión:", decision)
```

```python
prueba_rachas(datos)
```
